# Project dự đoán giá nhà

### 1. Vấn đề: Dự đoán giá nhà dựa trên các thuộc tính của căn nhà sao cho đúng nhất

```python
# import library
import pandas as pd
import numpy as np
```

```python
# read file train data
data = pd.read_csv("./data/train.csv", index_col = "Id")
```

```python
data.head()
```

```python
data.columns
```

### 2. Chọn lọc các đặc trưng (Feature selection)
Chọn lọc những cột đặc trưng nhất của data trong các cột

- `LotArea`: Diện tích lô đất
- `YearBuilt`: Năm xây dựng ngôi nhà
- `1stFlrSF`: Diện tích sàn tầng một
- `2ndFlrSF`: Diện tích sàn tầng hai
- `FullBath`: Số phòng tắm
- `BedroomAbvGr`: Số phòng ngủ đạt tiêu chuẩn
- `TotRmsAbvGrd`: Tổng số phòng


```python
features = ["LotArea", "YearBuilt", "1stFlrSF", "2ndFlrSF", "FullBath", "BedroomAbvGr", "TotRmsAbvGrd"]
```

### 3. Phân rã data
- `data`: dataset
- `X`: data[features]
- `y`: biến mục tiêu `SalePrice`

chia X, y ra thành `X_train`, `Y_train`, `X_valid`, `y_valid`
- `X_train`, `y_train`: Dùng để train mô hình máy học
- `X_valid`, `y_valid`: Dùng để kiểm tra

```python
X = data[features]
y = data["SalePrice"]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=0)
```

```python
X_train.shape
```

```python
X.shape
```

```python
X_valid.shape
```

### 4. Training Model

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=1)
```

```python
dt_model.fit(X_train, y_train)
```

```python
y_pr = dt_model.predict(X_valid)
```

```python
# Kết quả dự đoán với data là tệp dữ liệu X_valid
y_pr
```

```python
# So sánh kết quả dự đoán so với giá trị thật
pd.DataFrame({'y': y_valid, 'y_pr': y_pr})
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(random_state=1) 
rf_model.fit(X_train, y_train)
```

```python
rf_y_pr = rf_model.predict(X_valid)
```

```python
pd.DataFrame({'y': y_valid.head(), 'y_pr': rf_y_pr[:5]})
```

```python
X_valid.head()
```

```python
# Dự đoán với 1 giá trị duy nhất
rf_model.predict([[20105, 1970, 2200, 500, 2, 3, 6]])
```

# 5. Đánh giá mô hình (model evaluation)
### Các metrics đánh giá
- Mean Absolute Error (MAE)
- Mean Squared Error (MSE)
- R-squared (R²)

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Đánh giá mô hình Decision Tree
mae = mean_absolute_error(y_valid, y_pr)
mse = mean_squared_error(y_valid, y_pr)
r2 = r2_score(y_valid, y_pr)
```
